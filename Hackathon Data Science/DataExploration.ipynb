```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PMAppDummydata.csv')
df.head()
```

```python
df_teamSize = df.groupby(by=['teamSize'])['teamSize'].size().reset_index(name='count')
display(df_teamSize)
```

```python
sns.barplot(x='teamSize', y='count', data= df_teamSize).set(title='Count by Team Size')
```

```python
df_importance = df.groupby(by=['importance'])['importance'].size().reset_index(name='count')
sns.barplot(x='importance', y='count', data= df_importance).set(title='Count by Importance')
```

```python
df_workload = df.groupby(by=['workload'])['workload'].size().reset_index(name='count')
display(df_workload)
sns.barplot(x='workload', y='count', data= df_workload).set(title='Count by Importance')
```

```python
corr = df.corr(); corr
```

```python
sns.lmplot(x ='budget', y ='teamSize',
          fit_reg = False, hue = 'workload',
          data = df)
 
# Show the plot
plt.show()
```
