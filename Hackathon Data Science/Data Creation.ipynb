```python
import numpy as np
import pandas as pd
import random
```

```python
df_temp = pd.read_csv('Project_Management_Data.csv')
df_temp.head()
```

```python
df = df_temp[['ProjectTitle']].rename( columns = {'ProjectTitle':'name'}); df
```

```python
def divisible_random(a,b,n):
    if b-a < n:
      raise Exception('{} is too big'.format(n))
    result = random.randint(a, b)
    while result % n != 0:
      result = random.randint(a, b)
    return result
# get a random int in the range 2 - 12, the number is divisible by 3
print(divisible_random(2,12,3))
```

```python
df['teamSize'] = np.random.randint(2,26,size=len(df))
df['budget'] = [divisible_random(10000,1000000,1000) for i in range(len(df))]
df['workload'] = np.random.choice(['light', 'manageable', 'heavy'], size=len(df))
df['importance'] = np.random.randint(1,11,size=len(df))
df['completionTime'] = [divisible_random(5,100,5) for i in range(len(df))]
df.head()
```
